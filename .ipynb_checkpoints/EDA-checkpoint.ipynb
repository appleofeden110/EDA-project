To understand the distribution of the continuous variables in this Dataset, it is most likely that if we compile all three doses that the participants of the dataset had done into one Plot or Graph, it will be confusing, and as such the distribution will be shown individually using histograms and KDE plots respectively. 

First let's see how it looks with the original data using FirstDose Variable and dividing it by areaName

It will be divided into 55 bins for more clarity


```python
sns.displot(data, x="FirstDose", hue="areaName", bins=55, kde=True)
```

As we can see, it is confusing and data cannot be visually understood reasonably.

To fix that, the logarithmic transformation will be used 

```python
data["fd_log"] = np.log(data["FirstDose"])
data["sd_log"] = np.log(data["SecondDose"])
data["td_log"] = np.log(data["ThirdDose"])
```

Now to individually show the distribution of each continuous variable

```python
sns.displot(data, x="fd_log", hue="areaName", bins=55, multiple="stack", kde=True)
```

```python
sns.displot(data, x="sd_log", hue="areaName", bins=55, multiple="stack", kde=True)
```

```python
sns.displot(data, x="fd_log", hue="areaName", bins=55, multiple="stack", kde=True)
```

From here we can see clear distinction between England and rest of listed areas. England the biggest region in the United Kingdom, which makes the fact that number of vaccinations in that area will be far greater predictable. 

Out of other regions listed, Scotland is the second in the number of vaccinations, Wales third and Northern Ireland has the least number of vaccinations in listed regions. 

However, there is another way of distributing data of First, Second and Third Doses

Date might be the factor in receiving the vaccinations in all countries, as first, second and third doses variables in its names imply that they might have a been recorded in different times. Let's review that possibility by plotting it onto the Histagram/KDE distribution plot again

For clarity's sake, I will put a year and a month into month/year format as a string into a new column in the DataFrame

B. 

```python
# FOR LATER
# created the numpy array to store the formatted date string values
formatted_values = np.array([])
# iterate through rows to receive the data and append it into np array
for _, row in data.iterrows():
    formatted_values = np.append(formatted_values,  f"{row["month"]}/{row["year"]:.0f}")
#create the column
data["month/year"] = formatted_values
data=data.drop(data[data["year"] == 0].index)

def find_zero_year(year, month) -> float:
    if year == 0 :
        if 0 < month <= 5:
            y = 2022
            return float(y)
        elif 9 <= month <= 12:
            y = 2021
            return float(y)
    else: return float(0)

for _, row in data.iterrows():
    true_year = find_zero_year(row["year"], row["month"])
    
    if true_year != 0:
        row["year"] = true_year
        
        
```

The error can be noticed in the data, as there is 0 year in the dataset. As per assumption from other values in year Series, it is indeed not a year 0, it is a fault in the Dataframe which will be fixed with following loop, in which it will be checked for 0 in subset of variable "Year" and overwritten with an appropriate year. Also, it will re-do the month/year column respectively.
